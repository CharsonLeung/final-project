```python
from datetime import datetime, timedelta
from io import StringIO
import time
ohlcv_engine = create_engine("postgresql+psycopg2://postgres:Admin1234$@localhost:5432/bc2504p")
metadata = MetaData()
yester_date = datetime.now().date() - timedelta(days=1)
desired_date_time = datetime(yester_date.year, yester_date.month, yester_date.day, 21, 30, 0)
period2 = round(desired_date_time.timestamp())
df_ohlcv = []

for i in range (0, len(df_sp500_symbols)):
  start_time = time.perf_counter()
  try:
    url = str("https://query1.finance.yahoo.com/v8/finance/chart/" + df_sp500_symbols.symbol[i].replace('.','-') + "?period1=1657237004&period2=" + str(period2) + "&interval=1d&events=history")

    # https://query1.finance.yahoo.com/v8/finance/chart/AAPL?period1=1657237004&period2=1752778462&interval=1d&events=history 
    headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
  }
    response = requests.get(url, headers=headers)

    df_ohlcv2 = pd.read_json(StringIO(response.text))
    df_ohlcv3 = (df_ohlcv2)['chart']['result'][0]['indicators']['quote'][0]
    df_symbol = (df_ohlcv2)['chart']['result'][0]['meta']['symbol']
    timestamp_loc = (df_ohlcv2)['chart']['result'][0]['timestamp']
    date_time = []
    
    for j in range (0, (len(timestamp_loc))):
      date_time.append(datetime.fromtimestamp(timestamp_loc[j]).strftime("%Y-%m-%d %H:%M:%S"))
      #print(date_time)
    df_ohlcv4 = pd.DataFrame({
      'symbol' : df_symbol,
      'low_price_of_the_day' : (df_ohlcv3)['low'],
      'previous_close_price' : (df_ohlcv3)['close'],
      'high_price_of_the_day' : (df_ohlcv3)['high'],
      'open_price_of_the_day' : (df_ohlcv3)['open'],
      'volume' : (df_ohlcv3)['volume'],
      'timestamp' : timestamp_loc,
      'date_time' : date_time
    })
    df_ohlcv.append(df_ohlcv4)
    df_pd_ohlcv = pd.concat(df_ohlcv, ignore_index=True)
    # print(df_pd_ohlcv)
    df_pd_ohlcv.to_sql('stock_ohlc_data', ohlcv_engine, index=False, if_exists="replace")
  except Exception as e:
    print(f"Error {e} occurred.")
  end_time = time.perf_counter()
  elapsed_time = end_time - start_time
  request_time = i + 1
  print(f"request_time={request_time}, request_stock={df_symbol}, elapsed_time={elapsed_time}")
```
