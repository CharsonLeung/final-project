```python
import pandas as pd
import finnhub

finnhub_company_profile = finnhub.Client(api_key="")


be_read = finnhub_company_profile.company_profile2(symbol='AAPL')

print(be_read)
print(type(be_read)['country'])
for i in ('country', 
          'currency', 
          'estimateCurrency',
          'exchange',
            'finnhubIndustry',
              'ipo',
                'logo',
                  'marketCapitalization',
                    'name',
                      'phone',
                        'shareOutstanding',
                          'ticker', 
                            'weburl'):
  if str(type((be_read)[i])) == "<class 'str'>":
    data_type = "String"
  else:
    data_type = "Float"
  print(f"private {data_type} {i};")
```
